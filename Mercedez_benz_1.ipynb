```python
#Experimental
```

We concatenate both dataframes because the categorical features in both the sets contain some differrent values and label encoding them seperately may cause unseen issues.

```python
X_both = pd.concat([X_train, X_test])
X_both
```

```python
#Label encoder
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
```

```python
for col in X_both.select_dtypes(include=['object']):
    X_both[col] = le.fit_transform(X_both[col])
X_both.head()
```

```python
X_train_le = X_both[:4209]
X_train_le.head()
```

```python
X_test_le = X_both[4209:]
X_test_le.head()
```

```python
from sklearn.decomposition import PCA
```

```python
# We retain more than 99 percent of the Explained variance
pca = PCA(0.99, random_state=42)
Pca_train = pca.fit_transform(X_train_le)
Pca_test = pca.transform(X_test_le)
```

```python
Pca_train.shape , Pca_test.shape
```

We reduced the number of features from 364 to 24 while retaining more than 99% of explained variance

```python
import xgboost as xgb
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb_model = xgb.XGBRegressor()
```

```python
params = {'learning_rate' : [0.025,0.05,0.1],
         'max_depth' : [2,3,4],
          'gamma' : [0,0.01,0.02]
         }
```

```python
clf = GridSearchCV(xgb_model, params, scoring ='r2', verbose = 2 , n_jobs = -1)
clf.fit(X_train_le, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
xgb_model = xgb.XGBRegressor(params = {'gamma': 0, 'learning_rate': 0.05, 'max_depth': 3})
```
